```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
## Load the Trained Mode , Scaler pickle  , onehot
model = load_model('model.h5')

#Load the encoder and Scaler

with open('label_encoder_gender.pkl', 'rb') as file:
    label_encoder_gender = pickle.load(file)

with open('one_hot_encoder_geo.pkl', 'rb') as file:
    one_hot_encoder_geo = pickle.load(file)

with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)
```

```python
#Example Input Data
input_data = {
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000,
}
```

```python
geo_encoded = one_hot_encoder_geo.transform(
    [[input_data['Geography']]]
).toarray()

geo_encoded_df = pd.DataFrame(
    geo_encoded,
    columns=one_hot_encoder_geo.get_feature_names_out(['Geography'])
)
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
input_df['Gender']  = label_encoder_gender.transform([input_data['Gender']])[0]
```

```python
input_df = pd.concat([input_df.drop('Geography', axis=1) ,geo_encoded_df] , axis = 1)
```

```python
input_df
```

```python
input_scaled = scaler.transform(input_df)
input_scaled
```

```python
input_scaled.shape
```

```python
prediction = model.predict(input_scaled)
```

```python
prediction_prob = prediction[0][0]
prediction_prob
```

```python
label = 1 if prediction_prob >= 0.5 else 0

print("Probability:", prediction_prob)
print("Predicted Class:", label)
```
